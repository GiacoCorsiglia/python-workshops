# Getting Started with Plotting

To create plots in Python we're going to use `matplotlib`.  Matplotlib is a package that enables you to create all sorts of graphs/plots/figures in Python.  It's quite powerful and flexible.

To use Matplotlib you must import it at the top of your Python file/Jupyter notebook.  Usually, you'll do that like this:

````python
import matplotlib.pyplot as plt

# We'll also import numpy:
import numpy as np
````

Once you run that cell, you can access Matplotlib's functions by typing `plt.<name of function>`.

Matplotlib is a big package with many features.  For most plotting tasks, you will use the subset of Matplotlib known as `pyplot`.  That's what we imported above (and then we shortened it to just `plt` to make it easier to for us to type).  

You can (and will often) refer to the [Matplotlib documentation](https://matplotlib.org/contents.html) when plotting. I usually prefer to look just at the [pages about Pyplot](https://matplotlib.org/api/pyplot_summary.html).

## Real quick Matplotlib/Pyplot essentials

When using Pyplot, the basic function is simply called `plot`.  You write:

```python
plt.plot(x values here, y values here) 
```

For both the x values and y values, you pass (one dimensional!) Lists/numpy arrays of numbers:

````python
plt.plot([1, 2], [0.4, 1.4])

# This line prevents Matplotlib from clipping the y-axis:
plt.ylim(0)
plt.xlabel("x-axis")
````

## Interlude about learning how to program

When learning to program, try to construct mental models of what your code is doing, and why it works.  Developing this skill is, in my opinion, the essential task of learning to program—and also the fun part!  Of course you have to learn the syntax and vocabulary of your programming language (e.g., memorizing that `len()` gives you the length of a List) along the way, but the more robust your mental models are, the less memorization you have to do, and the more you can productively use Google/Stack Overflow effectively.

It's easy to fall into the mode of copy-pasting the magic encantations from some random website that seems to make your code do what you want.  This approach often works—and everyone does it, so no shame!  But you may run into trouble when (a) you need to do something slightly different from the code you found; (b) you need to make it work with your other code (which maybe you copied from elsewhere).

Whenever I copy-paste code, I try to make sure I understand:

0. The syntax (e.g., why are those parenthesis there?)
0. What kind of statement(s) am I copying (e.g., are they variable definitions? Function calls?)
0. The source of the "definitions" (e.g., what package does this variable/function come from?  Did I import that?)
0. What each of the function calls are supposed to do (e.g., `np.array(...)` is supposed to create an array; `plt.plot(...)` is supposed to create a plot)
0. What each function _parameter_ means (e.g., `plt.plot(x data here, y data here)`)
0. What the code is doing _behind the scenes_

I don't always make it all the way down the list!  And of course the last one is a rabbit hole.  But—I think if you get in the habit of doing this and if you hone your mental models (which this helps with), you'll find that programming becomes quickly less frustrating and much more powerful.

## Examples

````python
# Let's plot sin(x)
x = np.linspace(0, 5, 100)
y = np.sin(x)

plt.plot(x, y)
````

````python
# sin(x) and cos(x) on the same graph, with a legend
x = np.linspace(0, 5, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")

plt.legend()
````
